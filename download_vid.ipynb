```python
import os
import yt_dlp

video_path = os.path.join("audio", "%(title)s.%(ext)s")
url = "https://youtu.be/3NNcdq1CtX8"
# url = input("🎥 Enter YouTube URL: ")
# video_path = os.path.join("audio", "one")
# choice = input("Download as [1] MP3 (audio) or [2] MP4 (video)? ")

def download_video(url, video_path, video_format="mp3"):
    print(f"\nDownloading {url}")
    print("="*40)
    
    # ---

    # set param
    ydl_opts = {
        "format": "bestaudio/best",
        "outtmpl": video_path,
        "postprocessors": [{
            "key": "FFmpegExtractAudio",
            "preferredcodec": video_format,
            "preferredquality": "192",
        }],
    }

    # download file
    with yt_dlp.YoutubeDL(ydl_opts) as ydl:
        main = ydl.download([url])

    # 
    final_path = video_path + "." + video_format
    print(f">>>>> Download finished in {final_path}")
    
    return final_path

print(download_video(url, video_path))
```
